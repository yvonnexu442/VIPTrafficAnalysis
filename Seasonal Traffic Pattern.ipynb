```python
import sys
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import plotly.express as px
import plotly.graph_objects as go
import datetime, warnings
from IPython.display import Image
from IPython.display import display
from plotly.subplots import make_subplots
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
# dfSpring = pd.read_csv('0893332_Spring_Speed.csv')
# dfSummer = pd.read_csv('0893332_Summer_Speed.csv')
# dfFall = pd.read_csv('0893332_Fall_Speed.csv')
# dfWinter = pd.read_csv('0893332_Winter_Speed.csv')
# dfSpring = pd.read_csv('0893374_Spring_Speed.csv')
# dfSummer = pd.read_csv('0893374_Summer_Speed.csv')
# dfFall = pd.read_csv('0893374_Fall_Speed.csv')
# dfWinter = pd.read_csv('0893374_Winter_Speed.csv')
dfSpring = pd.read_csv('0899611_Spring_Speed.csv')
dfSummer = pd.read_csv('0899611_Summer_Speed.csv')
dfFall = pd.read_csv('0899611_Fall_Speed.csv')
dfWinter = pd.read_csv('0899611_Winter_Speed.csv')
monthListSpring = dfSpring['Month'].dropna().tolist()
monthListSummer = dfSummer['Month'].dropna().tolist()
monthListFall = dfFall['Month'].dropna().tolist()
monthListWinter = dfWinter['Month'].dropna().tolist()
dfSpring.head()
```

```python
dfSpringHr = dfSpring.groupby(['Time']).mean()
dfSummerHr = dfSummer.groupby(['Time']).mean()
dfFallHr = dfFall.groupby(['Time']).mean()
dfWinterHr = dfWinter.groupby(['Time']).mean()
dfSummerHr.columns.values
```

```python
dfTimeSpring = dfSpring[dfSpring['Time'] == 0].mean(axis=0).to_frame().T
dfTimeSummer = dfSummer[dfSummer['Time'] == 0].mean(axis=0).to_frame().T
dfTimeFall = dfFall[dfFall['Time'] == 0].mean(axis=0).to_frame().T
dfTimeWinter = dfWinter[dfWinter['Time'] == 0].mean(axis=0).to_frame().T
```

```python
timeList = [dfTimeSpring, dfTimeSummer, dfTimeFall, dfTimeWinter]
seasonList = [dfSpring, dfSummer, dfFall, dfWinter]
colName = ['Time', 'Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
for j in range(4):    
    for i in range(1, 24):
        temp = seasonList[j][seasonList[j]['Time'] == i].mean(axis=0).to_frame().T
        timeList[j] = timeList[j].append(temp, ignore_index=True)
    timeList[j] = timeList[j].drop(['Month'], axis=1)
    timeList[j] = timeList[j][colName]
```

```python
timeList[1].head()
```

```python
time = [int(i) for i in range(24)]
colName = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
```

```python
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=("Spring", 
                    "Summer", 
                    "Fall", 
                    "Winter"))
# colors = ['rgb(0, 0, 128)', 'rgb(15,82,186)', 'rgb(0,128,255)', 'rgb(0,142,204)', 'rgb(149,200,216)']
colors = ['rgb(242, 98, 107)', 'rgb(254, 186, 79)', 'rgb(137, 224, 119)', 'rgb(131, 195, 255)', 'rgb(195, 129, 253)', 
         'rgb(151, 57, 153)', 'rgb(195, 44, 87)']
for i in range(7):    
    fig.add_trace(go.Scatter(x=time, y=timeList[0][colName[i]],
                             mode='lines',
                             name=colName[i], 
                             legendgroup="group{}".format(i), 
                             line=dict(color=colors[i], width=3)),
                  row=1, col=1)

    fig.add_trace(go.Scatter(x=time, y=timeList[1][colName[i]],
                             mode='lines',
                             name=colName[i],
                             legendgroup="group{}".format(i),
                             line=dict(color=colors[i], width=3),
                             showlegend=False),
                  row=1, col=2)

    fig.add_trace(go.Scatter(x=time, y=timeList[2][colName[i]],
                             mode='lines',
                             name=colName[i],
                             legendgroup="group{}".format(i),
                             line=dict(color=colors[i], width=3),
                             showlegend=False),
                  row=2, col=1)

    fig.add_trace(go.Scatter(x=time, y=timeList[3][colName[i]],
                             mode='lines',
                             name=colName[i],
                             legendgroup="group{}".format(i),
                             line=dict(color=colors[i], width=3),
                             showlegend=False),
                  row=2, col=2)
for i in range(1,3):
    for j in range(1,3):  
        fig.update_xaxes(title_text="Hour in A Day", row=i, col=j)
        fig.update_yaxes(title_text="Average Traffic Speed", row=i, col=j)
fig.update_layout(title_text="Seasonal Traffic Speed For Sensor ID: 0899611",
                  autosize=False,
                  width=1000,
                  height=800,
                  margin=go.layout.Margin
                  (
                  l=10,
                  r=10,
                  b=100,
                  t=100,
                  pad=2
                  ))
fig.show()
```
