```python
import sys
import pandas as pd
import numpy as np
import re
import os
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import plotly.express as px
import plotly.graph_objects as go
import datetime, warnings
from IPython.display import Image
from IPython.display import display
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
rawdf = pd.read_csv("AvgWeekdaySpeedSample.csv")
display(rawdf[:12].ix[:,1:])
```

```python
IDList = list(rawdf.ix[:,0].dropna())
IDList = [int(i) for i in IDList]
IDList
```

```python
s1 = rawdf[:12].ix[:,1:]
ax = sns.lineplot(y='Mon', x='Month', data=s1)
```

```python
y_axis = list(rawdf.columns.values.tolist())[2:]
y_axis
```

```python
plt.figure(figsize=(10,5))
plt.plot(s1[y_axis])
plt.legend(y_axis)
plt.title("Average Speed Per Month for Sensor {}".format(IDList[0]))
plt.xlabel("Month")
plt.ylabel("Speed (miles/hr)")
plt.show()
```

```python
start = 0
end = 12
counter = 0
while end < rawdf.shape[0]:
    s = rawdf[start:end].ix[:,1:]
    plt.figure(figsize=(10,5))
    plt.plot(s[y_axis])
    plt.legend(y_axis, bbox_to_anchor=(1.01, 1), loc='upper left')
    plt.title("Average Speed Per Month for Sensor {}".format(IDList[counter]))
    plt.xlabel("Month")
    plt.ylabel("Speed (miles/hr)")
    start += 12
    end += 12
    counter += 1
```
